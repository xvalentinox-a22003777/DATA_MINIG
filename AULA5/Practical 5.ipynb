#### Question 9: 
Create two bins with the names "Low" if the value is lower than 20000 and "High" if it's bigger than 20000 and less than 60000. Group the column "price" according to the bins and add a new column (price_range) to the dataset with the result.    

Save the cleaned dataset in "Automobile_data_cleaned.csv"

```python
bins = [0, 20000, 60000]
labels = ['Low', 'High']

df['price_range'] = pd.cut(df['price'], bins=bins, labels=labels, right=False)

df.to_csv('Automobile_data_cleaned.csv', index=False)
```
